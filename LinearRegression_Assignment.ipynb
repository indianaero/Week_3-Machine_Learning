```python
import pandas as pd # importing the pandas library for importing the dataset from csv files
```

```python
dataset = pd.read_csv("Salary_Data.csv")  #Using Pandas reading the csv file and saving it in the variable dataset
```

```python
dataset # checking the output of the dataset variable and we have YearsExperience as input and Salary as output
```

```python
independent = dataset[["YearsExperience"]]
dependent = dataset[["Salary"]] # Splitting the input and output and assigning them to respective variables
```

```python
independent # verifying the independent variable output
```

```python
dependent # checking the dependent variable output
```

```python
# Next step is to split the dataset into training and testing sets, For that import the sklearn library
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test =train_test_split(independent, dependent, test_size=0.30, random_state= 0)
#Splitting the inputs and outputs into training and testing datasets and assigning them to respective variables
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() #Next step is the model creation for that importing the linear regression model
```

```python
regressor.fit(X_train, y_train) #Training the model on the training data inputs and outputs
```

```python
weight = regressor.coef_ #finding the weight/slope of the regression model
weight
```

```python
bias= regressor.intercept_ #finding the min value/bias/orgin of the regression model
bias
```

```python
y_predict= regressor.predict(X_test) # Using .predict pedicting the input test data - Prediction part starts here -after the learning part
```

```python
from sklearn.metrics import r2_score # After prediction , evaluating the prediction using R Square 
r_score = r2_score(y_test, y_predict) # cheking the prediction score by comparing acutal test outputs and predicted test outputs
```

```python
r_score # if score close to 1 , the model prediction is close to accurate
```

```python
import pickle
filename ="final_model_linear.sav" # importing the pickle package to save the model as .sav file 
```

```python
pickle.dump(regressor,open(filename,"wb")) # Model is added to the file and saved as write - binary file
```

```python
load_model=pickle.load(open(filename,"rb")) # Deployment step - Open the saved model - should be in separate notebook so that end users can access only this and the steps below.
```

```python
result= load_model.predict([[15]])# predict using the model for new input
```
